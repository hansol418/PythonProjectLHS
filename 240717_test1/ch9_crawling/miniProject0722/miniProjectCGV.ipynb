```python
import time
import bs4
import urllib.request
import pymysql
import re

def clean_text(inputString):
    text_rmv = re.sub(
        '[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', inputString)
    return text_rmv

def insertData(title, price, cardimgUrl, productlink):
    try:
        con = pymysql.connect(host='127.0.0.1', user='root',
                              password='123456', database='pydb', charset='utf8')
        cur = con.cursor()
        query = "INSERT INTO CGVTable (title, price, cardimgUrl, productlink) VALUES (%s, %s, %s, %s)"
        values = (title, price, cardimgUrl, productlink)
        cur.execute(query, values)
        con.commit()
    except Exception as e:
        print(f"예외 발생: {e}")
    finally:
        con.close()

def main():
    count = 1
    nateUrl = "http://www.cgv.co.kr/movies/?lt=1&ft=0"

    while count < 201:
        try:
            htmlObject = urllib.request.urlopen(nateUrl)
            webPage = htmlObject.read()
            bsObject = bs4.BeautifulSoup(webPage, 'html.parser')

            # 'ol' 태그 안에 있는 'li' 태그들을 찾습니다
            ol_tags = bsObject.find_all('ol')
            for ol in ol_tags:
                li_tags = ol.find_all('li')
                for tag in li_tags:
                    if count >= 201:
                        break

                    # 제목 추출
                    title_tag = tag.find('strong', {'class': 'title'})
                    title = title_tag.text.strip() if title_tag else '제목 없음'
                    title = clean_text(title)

                    # 가격/점수 추출
                    price_tag = tag.find('strong', {'class': 'percent'})
                    price = price_tag.text.strip() if price_tag else '정보 없음'
                    price = clean_text(price)

                    # 이미지 URL 추출
                    imgLink = tag.find('span', {'class': 'thumb-image'})
                    if imgLink:
                        img_tag = imgLink.find('img')
                        cardimgUrl = img_tag['src'] if img_tag and 'src' in img_tag.attrs else '이미지가 존재하지 않음'
                    else:
                        cardimgUrl = '이미지가 존재하지 않음'

                    # 상품 링크 추출
                    product_link_tag = tag.find('a', href=True)
                    productlink = 'http://www.cgv.co.kr' + product_link_tag['href'] if product_link_tag else '링크 없음'

                    # 데이터 삽입
                    print("순서1: 데이터 추가 전")
                    print("title: %s" % (title))
                    print("price: %s" % (price))
                    print("productlink: %s" % (productlink))
                    print("cardimgUrl: %s" % (cardimgUrl))
                    print('================================')
                    print('================================')
                    insertData(title, price, cardimgUrl, productlink)
                    count += 1

                    if count >= 201:
                        break

            time.sleep(1)
            print(f"count : {count}")

        except Exception as e:
            print(f"오류 발생: {e}")
            break

    print("완료!")

if __name__ == "__main__":
    main()
```
